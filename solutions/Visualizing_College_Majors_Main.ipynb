## Visualizing College Majors Data
---

A dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

 - Rank - Rank by median earnings (the dataset is ordered by this column).
 - Major_code - Major code.
 - Major - Major description.
 - Major_category - Category of major.
 - Total - Total number of people with major.
 - Sample_size - Sample size (unweighted) of full-time.
 - Men - Male graduates.
 - Women - Female graduates.
 - ShareWomen - Women as share of total.
 - Employed - Number employed.
 - Median - Median salary of full-time, year-round workers.
 - Low_wage_jobs - Number in low-wage service jobs.
 - Full_time - Number employed 35 hours or more.
 - Part_time - Number employed less than 35 hours.

Using visualizations, we can start to explore questions from the dataset like:

- Do students in more popular majors make more money?
    - Using scatter plots
- How many majors are predominantly male? Predominantly female?
    - Using histograms
- Which category of majors have the most students?
    - Using bar plots

We'll explore how to do these and more while primarily working in pandas. 

### Exploring Dataset

```python
import pandas as pd
import matplotlib as plt
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
#recent_grads.tail()
```

Dropping null values

```python
raw_data_count = len(recent_grads)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
print('Initial num of rows: {}\nCleaned num of rows: {}'.format(raw_data_count,cleaned_data_count))
```

### Scatterplots

Using scatter plots, visualize following relations:
 - Sample_size and Median
 - Sample_size and Unemployment_rate
 - Full_time and Median
 - ShareWomen and Unemployment_rate
 - Men and Median
 - Women and Median

#### Sample Size to Median Relation

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter',
                title='Sample_size to Median Relation', figsize=(5,10))
```

#### Sample_size to Unemployment_rate relation

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter',
                title='Sample_size to Unemployment_rate Relation', figsize=(5,10))
```

#### Full_time to Median Relation

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter',
                title='Full_time to Median Relation', figsize=(10,10))
```

#### Men to Median Relation

```python
recent_grads.plot(x='Men', y='Median', kind='scatter',
                title='Men to Median Relation', figsize=(10,10))
```

#### Women to Median Relation

```python
recent_grads.plot(x='Women', y='Median', kind='scatter',
                title='Women to Median Relation', figsize=(10,10))
```

### Findings
---
##### Do students in more popular majors make more money?
More popular majors tend to have their median salary close to the bottom line. Therefore students in those majors are likely to have lower salary than others.

##### Do students that majored in subjects that were majority female make more money?
Majors with the majority of women are likely to have lower wage than others. Medium salary in those groups a close to the minimum.

##### Is there any link between the number of full-time employees and median salary?
Can't assume that based on the plots. As it requires to compute relative values.

### Bar Plots

Use bar plots to compare the percentages of women (ShareWomen) from the 10 highest paying majors and from the 10 lowest paying majors.

```python
recent_grads.sort_values('Median',ascending=False)[:10].plot.bar(x='Major',y='ShareWomen')
```

```python
recent_grads.sort_values('Median')[:10].plot.bar(x='Major',y='ShareWomen')
```

Use bar plots to compare the unemployment rate (Unemployment_rate) from the 10 highest paying majors and from the 10 lowest paying majors.

```python
recent_grads.sort_values('Median', ascending=False)[:10].plot.bar(x='Major',y='Unemployment_rate')
```

```python
recent_grads.sort_values('Median')[:10].plot.bar(x='Major',y='Unemployment_rate')
```

### Findings

Top 10 highest salary majors have low women employment rate.Except Astrophysics.

Top 10 lowest salary majors are presented by women mostly (80-90 percetage rate)

Nuclear Engineers have highest unemployment rate among top salary majors - 18%.

As for bottom-salary-majors, clinical psychology is has highest unemployment rate among others - 14%.

## Visualize degree categories

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

### All Degrees
---
Visualizing all degrees in the dataset, grouped into categories.

Each column represents a category: STEM, liberal arts and other.

```python
fig, axes = plt.subplots(6,3, figsize=(18,24))

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

#Building STEM and Health Others charts
for pos in range(0,6):
    #STEM plots
    stem_ax = axes[pos,0]
    stem_ax.plot(women_degrees['Year'], women_degrees[stem_cats[pos]], c=cb_dark_blue, label='Women', linewidth=3)
    stem_ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[pos]], c=cb_orange, label='Men', linewidth=3)
    stem_ax.set_title(stem_cats[pos])
    
    #Others plots
    others_ax = axes[pos,2]
    others_ax.plot(women_degrees['Year'], women_degrees[other_cats[pos]], c=cb_dark_blue, label='Women', linewidth=3)
    others_ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[pos]], c=cb_orange, label='Men', linewidth=3)
    others_ax.set_title(other_cats[pos])
    
    #Liberal Arts plots
    if pos <5:
        lib_ax = axes[pos,1]
        lib_ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[pos]], c=cb_dark_blue, label='Women', linewidth=3)
        lib_ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[pos]], c=cb_orange, label='Men', linewidth=3)
        lib_ax.set_title(lib_arts_cats[pos])
    
    #List of category axes:
    list_of_axes = [ax for axgroup in axes for ax in axgroup]
    
    for ax in list_of_axes:
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')

#Text annotations for STEM. First and last plots.
axes[0,0].text(2005,85,'Women')
axes[0,0].text(2005,16,'Men')
axes[5,0].text(2005,7,'Women')
axes[5,0].text(2005,90,'Men')

#Text annotations for Liberal Arts. First plot.
axes[0,1].text(2005,80,'Women')
axes[0,1].text(2005,18,'Men')

#Text annotations for Others. First and last plots.
axes[0,2].text(2005,90,'Women')
axes[0,2].text(2005,7,'Men')
axes[5,2].text(2005,30,'Women')
axes[5,2].text(2005,70,'Men')

#Add label bottom to the bottom of each group
axes[5,0].tick_params(labelbottom='on')
axes[4,1].tick_params(labelbottom='on')
axes[5,2].tick_params(labelbottom='on')

#Set y labels and horizontal line
for ax in list_of_axes:
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255, 171/255, 171/255), alpha=0.3)
```
